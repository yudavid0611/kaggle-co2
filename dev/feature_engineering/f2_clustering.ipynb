# f2. Clustering

```python
import os
import pandas as pd
```

```python
DATA_PATH = '../../data/'
train = pd.read_csv(os.path.join(DATA_PATH, 'train_f1.csv'), index_col=0)
test = pd.read_csv(os.path.join(DATA_PATH, 'test_f1.csv'), index_col=0)

# train_origin = train.copy()
# test_origin = test.copy()

train.head()
```

## 1) Feature selection for clustring

```python
features = ['latitude', 'longitude', 'week_no']

X = train.copy()
X = X.loc[:, features]
X_scaled = (X - X.mean(axis=0)) / X.std(axis=0)
```

## 2) K-Means Clustering

```python
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
# K-Means Clustering 수행
kmeans = KMeans(n_clusters=8, random_state=0)
X["Cluster"] = kmeans.fit_predict(X_scaled)


X.head(10)
```

```python
train['cluster'] = X['Cluster']
```

```python
# 데이터 저장
data_name = os.path.join(DATA_PATH, 'train_f2.csv')
train.to_csv(data_name)
```

## 3) Test data

```python
X_test = test.copy()
X_test = X_test.loc[:, features]
X_test_scaled = (X_test - X_test.mean(axis=0)) / X_test.std(axis=0)
```

```python
X_test_scaled
```

```python
X_test_scaled['Cluster'] = kmeans.predict(X_test_scaled)
X_test_scaled.head()
```

```python
test['cluster'] = X_test_scaled['Cluster']
```

```python
# 데이터 저장
data_name = os.path.join(DATA_PATH, 'test_f2.csv')
test.to_csv(data_name)
```
