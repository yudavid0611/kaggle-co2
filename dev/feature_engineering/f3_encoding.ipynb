# f3. Encoding

```python
import os
import pandas as pd
```

```python
DATA_PATH = '../../data/'
train = pd.read_csv(os.path.join(DATA_PATH, 'train_f2.csv'), index_col=0)
test = pd.read_csv(os.path.join(DATA_PATH, 'test_f2.csv'), index_col=0)

# train_origin = train.copy()
# test_origin = test.copy()

train.head()
```

## 1) One-hot encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# sparse=False: 인코딩된 칼럼이 numpy array type을 갖도록
encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
```

```python
train['cluster'] = train['cluster'].astype('O')
test['cluster'] = test['cluster'].astype('O')
```

```python
oh_train = pd.DataFrame(encoder.fit_transform(train[['cluster']]))
oh_test = pd.DataFrame(encoder.transform(test[['cluster']]))

# 인덱스 되돌리기
oh_train.index = train.index
oh_test.index = test.index
```

```python
train.drop('cluster', axis=1, inplace=True)
test.drop('cluster', axis=1, inplace=True)
```

```python
train = pd.concat([train, oh_train], axis=1)
test = pd.concat([test, oh_test], axis=1)
```

```python
# train 저장
data_name = os.path.join(DATA_PATH, 'train_f3.csv')
train.to_csv(data_name)
```

```python
# test 저장
data_name = os.path.join(DATA_PATH, 'test_f3.csv')
test.to_csv(data_name)
```
