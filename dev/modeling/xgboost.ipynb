## 1) 데이터

```python
import os
import pandas as pd
```

```python
DATA_PATH = '../../data/'
train = pd.read_csv(os.path.join(DATA_PATH, 'train_f2.csv'), index_col=0)
# test = pd.read_csv(os.path.join(DATA_PATH, 'test.csv'))

train_origin = train.copy()
# test_origin = test.copy()

train.head()
```

## 2) Feature selection and split dataset

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
train.drop(['ID_LAT_LON_YEAR_WEEK', 'year'], axis=1, inplace=True)
train.head()
```

```python
X = train.drop('emission', axis=1)
y = train['emission']

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, random_state=0)
```

## 3) Modeling

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
```

```python
# set up our search grid
param_grid = {"max_depth":    [4, 5, 6, 7],
              "n_estimators": [500, 700, 900, 1100],
              "learning_rate": [0.01, 0.001]}
```

```python
regressor=xgb.XGBRegressor(eval_metric='rmse')
```

```python
# try out every combination of the above values
search = GridSearchCV(regressor, param_grid, cv=5).fit(X_train, y_train)
print("The best hyperparameters are ", search.best_params_)
```
